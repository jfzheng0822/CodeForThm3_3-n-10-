```python
for l in [5..9]: 
    T=graphs(l);
    for G in T:
        if G.is_connected():
            if G.is_bipartite():
                if G.diameter()==4:
                    DD=G.distance_matrix(); 
                    E=DD.eigenvalues(); 
                    E1=sorted(E);
                    if E1[3]<0 and E1[4]>=0: 
                        G.show();
                        print(E1);
```
